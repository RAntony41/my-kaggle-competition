```python
import json
```

```python
# Load the JSON file into a dictionary
with open('capgen_v1.0_val.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
data
```

```python
path = []
label = []
for i in data.keys():
    temp = i
    
    for j in data[i]:
        
        if i[:12] == 'coco/val2017':
            temp = '/project/lt900502-ck24tn/captioning/coco/val2017' + i[-17:]
        elif i[:14] == 'coco/train2017':
            temp = '/project/lt900502-ck24tn/captioning/coco/train2017' + i[-17:]
            
        elif i[:16] == 'ipu24/val/travel':
            temp = '/project/lt900502-ck24tn/captioning/ipu24/val/travel' + i[-10:]
        elif i[:14] == 'ipu24/val/food':
            temp = '/project/lt900502-ck24tn/captioning/ipu24/val/food' + i[-10:]
        
        elif i[:18] == 'ipu24/train/travel':
            temp = '/project/lt900502-ck24tn/captioning/ipu24/train/travel' + i[-10:]
        elif i[:16] == 'ipu24/train/food':
            temp = '/project/lt900502-ck24tn/captioning/ipu24/train/food' + i[-10:]
        
        path.append(temp)
        label.append(j)
```

```python
len('00000.jpg')
```

```python
path
```

```python
final_dic = []

for i, _ in enumerate(label):
  dic = {
    "id": str(i),
    "image": [path[i]],
    "conversations": [
      {
        "from": "user",
        "value": "<ImageHere> Please describe this images in Thai."
      },
      {
        "from": "assistant",
        "value": label[i]
      }
    ]
    }
  final_dic.append(dic)
```

```python
final_dic
```

```python
with open('llamaformat_capgen_v1.0_val.json', 'w', encoding='utf-8') as file:
    json.dump(final_dic, file, ensure_ascii=False, indent=4)
```
