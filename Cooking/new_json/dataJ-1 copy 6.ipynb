```python
def input_output(path):
    # Initialize an empty dictionary
    data_dict = {}

    # Open the file using with statement
    with open(path, 'r') as file:
        # Read the entire content of the file
        content = file.read()
        # Split the content into parts based on the known structure
        delimiter_pattern = r"(?:\nOutput Script:\n|\nOutput Script: \n|\nOutput Script: )"

        # Split the content using the regular expression
        parts = re.split(delimiter_pattern, content)

        # Assign the parts to the dictionary with appropriate keys
        data_dict['input'] = parts[0].replace('Task: ', '').strip()
        if (parts[1].strip())[:16] == 'pyautogui.click(':
            data_dict['output'] = (parts[1].strip())[16:]
        elif (parts[1].strip())[:17] == 'npyautogui.click(':
            data_dict['output'] = (parts[1].strip())[17:]
        
    # Extract the numbers from the 'Output Script' value
    data_dict['output'] = re.findall(r"[-+]?\d*\.\d+|\d+", data_dict['output'])

    # Convert the extracted numbers to float
    data_dict['output'] = [int(float(data_dict['output'][0])/2.9),int(float(data_dict['output'][1])/1.8)] 
    return data_dict
```
