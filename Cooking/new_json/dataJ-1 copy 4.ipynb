```python
instruction = [str(i) for i in ds.data['train']['instruction']]
input = [str(i) for i in ds.data['train']['input']]
output = [str(i) for i in ds.data['train']['output']]
```

```python
final_dic = []
id = 0
for i in range(len(output)):
  user = ''
  assis = ''
  '''
  for j in i:
    if j['role'] == 'user':
      user = j['content']
    elif j['role'] == 'assistant':
      assis = j['content']
  ''' 
  dic = {
    "id": str(i),
    "conversations": [
      {
        "from": "user",
        "value": input[i] + instruction[i]
      },
      {
        "from": "assistant",
        "value": output[i]
      }
      ]
      }
  id = id + 1
  final_dic.append(dic)
```
