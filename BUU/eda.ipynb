```python
import pandas as pd
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('./hackathon-online-data-science-mini-epidemiology/Train_set.csv')
test = pd.read_csv('./hackathon-online-data-science-mini-epidemiology/Test_set.csv')
sample_submission = pl.read_csv('./hackathon-online-data-science-mini-epidemiology/sample_submission.csv')
```

```python
train
```

```python
plt.figure(figsize=(15, 10))

# Loop through each column and plot its distribution
for i, col in enumerate(train.columns):
    plt.subplot(len(train.columns)//2 + 1, 2, i+1)
    sns.histplot(train[col], kde=True)
    plt.title(col)

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
