```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch
from chronos import ChronosPipeline
```

```python
pipeline = ChronosPipeline.from_pretrained(
  "amazon/chronos-t5-tiny",
  torch_dtype=torch.bfloat16,
)
```

```python
train = pd.read_csv('/kaggle/input/mitrphol-ai-hackathon/data_collection.csv')
train
```

```python
test = pd.read_csv('/kaggle/input/mitrphol-ai-hackathon/sandbox_submission.csv')
test
```

```python
import matplotlib.pyplot as plt
```

```python
train['sugar_price'].plot()
plt.show()
```

```python
context = torch.tensor(train["sugar_price"])
prediction_length = 365
forecast = pipeline.predict(context,prediction_length,limit_prediction_length=False)
```

```python
forecast.shape
```

```python
averaged_forecast = (torch.mean(forecast, dim=1).numpy())[0]
```

```python
pd.Series(averaged_forecast).plot()
plt.show()
```

```python
pred = averaged_forecast
```

```python
pred.shape
```

```python
test['Prediction'] = pred
```

```python
test
```

```python
test['Prediction'].plot()
plt.show()
```

```python
test.to_csv('submission.csv',index=False)
```
